# Scatter Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Run the line below to download the dataset

```python
gdp_df = pd.read_csv('4_ThreeNum.csv')
gdp_df.head()
```

```python
gdp_df.describe()
```

## Part 1

```python
fig = plt.figure(figsize=(10, 5))
ax = sns.scatterplot(
    data=gdp_df,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    hue = "continent",
    sizes=(20, 2000),
    alpha = 0.5,
    palette = "deep"
)
ax.legend(bbox_to_anchor=(0.5, -0.3), loc = "lower center", ncol=5)
ax.spines[:].set_visible(False)
# handles, labels = ax.get_legend_handles_labels()
ax.grid(True)

plt.show()
```
